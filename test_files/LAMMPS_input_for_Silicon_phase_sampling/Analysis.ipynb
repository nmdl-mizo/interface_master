```python
x_coordinates = np.unique(data[:,0])
```

```python
low_index = []
for i in range(len(x_coordinates)):
    index_at_x_data = np.where(data[:,0] == x_coordinates[i])[0]
    lowest_energy_at_x = min(data[:,7][index_at_x_data])
    index_at_x_data_min = np.where(data[:,7][index_at_x_data] == \
                                  lowest_energy_at_x)[0]
    index = index_at_x_data[index_at_x_data_min]
    if len(index > 1):
        index = index[0]
    low_index.append(index)
    if i == 0:
        lowest_data = data[index]
    else:
        lowest_data = np.vstack((lowest_data, data[index]))
```

```python
lens = [len(data_AA),len(data_BB),len(data_AB)]
names = ['work_AA','work_BB','work']
low_index_normal = []
low_name = []
for i in range(len(low_index)):
    here = low_index[i]
    found = False
    count = 0
    while not found:
        if here - lens[count] > 0:
            here = here - lens[count]
            count += 1
        else:
            found = True
            low_index_normal.append(here + 1)
            low_name.append(names[count])
```

```python
low_information = np.column_stack((low_name, low_index_normal))
```
