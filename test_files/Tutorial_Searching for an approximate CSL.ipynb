```python
from interfacemaster.cellcalc import get_primitive_hkl
from interfacemaster.interface_generator import core, print_near_axis, convert_vector_index
from numpy import array
from numpy import array, dot, round
from numpy.linalg import inv
```

# Define an interface core class

# Input two cif files.

For GBs, input the same file; for interfaces, input two different files.


The first structure is non-changed while the second one can be deformed to form an approximate CSL

Please input files of CONVENTIONAL cells if you are not familiar with the indices in the primitive cell.

```python
my_interface = core('cif_files/Si_mp-149_conventional_standard.cif', \
                    'cif_files/SiC_mp-568656_conventional_standard.cif')
```

Find two vectors from different crystals with close length

```python
print_near_axis(0.1, my_interface.lattice_1, my_interface.lattice_2)
```

```python
axis = convert_vector_index(my_interface.conv_lattice_1, my_interface.lattice_1, [1,-1,1])
axis
```

We select [ 4 -2 -2 ] of lattice 1, [ 2 -3 0 ] of lattice 2.

Adjust the orientation of lattice 2 to make these two vectors colinnear 

```python
my_interface.set_orientation_axis([4, -2, -2], [2, -3, 0])
```

Searching for an approximate CSL by rotating crystal 2 along the [4, -2, -2] axis of
crystal 1

```python
my_interface.parse_limit(du = 0.2, S  = 0.2, sgm1=100, sgm2=100, dd = 0.2)
#Do searching... this might take a long time...
axis = [4, -2, -2]
my_interface.search_one_position(axis,0,180,0.5)
```

# Determining the interface plane to be a CSL lattice plane

```python
hkl = get_primitive_hkl(array([1,0,0]), my_interface.CSL, my_interface.lattice_1)
hkl
```

# Make a bicrystal!
Please refer to another tutorial to get the meaning of the relative parameters 

```python
my_interface.compute_bicrystal(hkl, orthogonal = False)
my_interface.get_bicrystal(xyz_1 = [1,1,1], xyz_2 = [1,1,1], dp2 = 1)
```

```python
CNID_conv = round(dot(inv(my_interface.conv_lattice_1)\
                              , my_interface.CNID),8)
CNID_conv
```
